### 1) Take two 3D coordinates from the user. Find out the distance between these points using following distance functions.x and y are vectors representing the coordinates of two points.Here, k=3

```python
import numpy as np

# Function to calculate Euclidean distance
def euclidean_distance(x, y):
    return np.sqrt(np.sum((x - y)**2))

# Function to calculate Manhattan distance
def manhattan_distance(x, y):
    return np.sum(np.abs(x - y))

# Function to calculate Minkowski distance
def minkowski_distance(x, y, p):
    return np.power(np.sum(np.abs(x - y)**p), 1/p)

# Take input coordinates from the user
li1 = list(map(int,input("Enter x values: ").split()))
li2 = list(map(int,input("Enter y values: ").split()))
x = np.array(li1)

y = np.array(li2)

# Calculate distances
euclidean_dist = euclidean_distance(x, y)
manhattan_dist = manhattan_distance(x, y)
minkowski_dist = minkowski_distance(x, y, 3)

# Print distances
print(f"Euclidean Distance: {euclidean_dist}")
print(f"Manhattan Distance: {manhattan_dist}")
print(f"Minkowski Distance (k=3): {minkowski_dist}")
```

### 2) Import the Iris dataset. Write a program to obtain the Euclidian Distance Matrix for all the data samples in the feature space. Distance metric is a 2D array, where the (i,j)th entry represents the distance between the ith and jth sample points in the feature space.

```python
import numpy as np
from sklearn.datasets import load_iris
from scipy.spatial.distance import euclidean

# Step 1: Load the Iris dataset
iris = load_iris()
data = iris.data

# Step 2: Compute the number of samples and features
num_samples, num_features = data.shape

# Step 3: Initialize the Euclidean Distance Matrix
distance_matrix = np.zeros((num_samples, num_samples))

# Step 4: Compute pairwise Euclidean distances
for i in range(num_samples):
    for j in range(num_samples):
        if i != j:  # Exclude distance of a sample to itself
            distance_matrix[i, j] = euclidean(data[i], data[j])

# Step 5: Display the Euclidean Distance Matrix
print("Euclidean Distance Matrix:")
print(distance_matrix)
```

### 3)Import the Iris dataset. Prepare a dataset considering samples belong to any two output classes. Draw the scatter plot for all the samples in the new dataset considering any two input attributes. Examine the scatter plot to find the equation of a line that can separate sample of two classes.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Step 1: Load the Iris dataset
iris = load_iris()

# Step 2: Create a DataFrame with the data
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                  columns= iris['feature_names'] + ['target'])

# Step 3: Select two output classes
class_0 = 0  # Select one class
class_1 = 1  # Select another class

# Step 4: Prepare dataset for the selected classes
df_selected = df[(df['target'] == class_0) | (df['target'] == class_1)]

# Step 5: Select two input attributes
input_attr1 = 'sepal length (cm)'
input_attr2 = 'sepal width (cm)'

# Step 6: Scatter plot for the selected input attributes
plt.figure(figsize=(8, 6))
plt.scatter(df_selected[df_selected['target'] == class_0][input_attr1], 
            df_selected[df_selected['target'] == class_0][input_attr2], 
            color='red', label=f'Class {class_0}')
plt.scatter(df_selected[df_selected['target'] == class_1][input_attr1], 
            df_selected[df_selected['target'] == class_1][input_attr2], 
            color='blue', label=f'Class {class_1}')
plt.xlabel(input_attr1)
plt.ylabel(input_attr2)
plt.title(f'Scatter Plot of {input_attr1} vs {input_attr2} for Two Classes')
plt.legend()
plt.grid(True)
plt.show()
```
